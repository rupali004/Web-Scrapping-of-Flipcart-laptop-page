```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
import re
#web scrapping of flipkart page and get the information about laptops
r=requests.get('https://www.flipkart.com/q/best-laptops-under-rs-50000')
soup=BeautifulSoup(r.content,'html.parser')
print(soup.prettify())
```

```python
Items=soup.find_all('div',class_="_2kHMtA")
Items
```

```python
print(type(Items),len(Items))
```

```python
#total 24 products are available
Items[0]
```

```python

for Item in Items:
    product=Item.findAll('div',{'class':'_4rR01T'})
    print(product[0].text.split('-')[0])
  #To find all tags  
    
```

```python
f=open('Laptop_info.csv','wb')
f.write("productName,stars,Ratings,Reviews,current_price,MRP,processor,RAM,Storage,ImageURL\n".encode())

for Item in Items:
# finding product name
    product=Item.find('div',{'class':'_4rR01T'})
    productName=product.text.split('-')[0].strip()
    
  
    # To find all stars of laptop
    
    stars=Item.find('div',{'class':'_3LWZlK'})
    try:
        stars=stars.text
    except:
        stars=0
        
    # To find all ratings and Reviews of laptop
    Rating=Item.find('span',{'class':'_2_R_DZ'})
    try:
        ratRev=re.findall('\d+,?\d*',Rating.text)
        Ratings=ratRev[0].replace(',','')
        Reviews=ratRev[1].replace(',','')
    except:
        Ratings=0
        Reviews=0
    
    
    # To find current Rating
    current_price=Item.find('div',{'class':'_30jeq3 _1_WHN1'}).text.replace(',','')
   
    
    # To find MRP
    MRP=Item.find('div',{'class':'_3I9_wc _27UcVY'})
    try:
        MRP=MRP.text.replace(',','')
    except:
        MRP=0
    
    # To finding information of the products
    info=Item.findAll('li',{'class':'rgWa7D'})
    processor=info[0].text
    RAM=info[1].text
    Storage=info[3].text
   
    
    Image=Item.img
    ImageURL=Image.get('src')
    print(productName,stars,Ratings,Reviews,current_price,MRP,processor,RAM,Storage,ImageURL)
    f.write(f"{productName},{stars},{Ratings},{Reviews},{current_price},{MRP},{processor},{RAM},{Storage},{ImageURL}\n".encode())
    print('\n')
f.close()  
```
